```python
features = ['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio',' stroke','compressionratio','horsepower'
,'peakrpm','citympg']
x = df[features]
y = df['price']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=34,test_size=0.1)
y_test.shape
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test,y_test)
print('Akurasi model Regresi Linear', score)
```

```python
input_data = np.array([[1,2,88.6,168.8,64.1,48.8,2548,130]])

prediction = lr.predict(input_data)
print('Estimasi Car Price :',prediction)
print('Nilai Car price : 6')
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('R2:', r2_score(y_test,pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, pred)))
```
